```python
## WAP that creates a DF from a list of dictionary

import numpy as np
import pandas as pd

Data=[{'Name':'Navya','Marks':50},{'Name':'Kiran','Marks':90},{'Name':'Ram','Marks':100}]
df=pd.DataFrame(Data,index=[1,2,3])
print(df)
```

```python
## WAP that creates a DF from a list of dictionary

import numpy as np
import pandas as pd

Data=[{'Name':'Navya','Marks':50,'Age':20},{'Name':'Kiran','Marks':90,'Age':30},{'Name':'Ram','Marks':100,'Age':32}]
df=pd.DataFrame(Data,index=[1,2,3])
print(df)
```

```python
## WAP to create a DF from a dictionary of series


import numpy as np
import pandas as pd

Data={'Karan':pd.Series([89,80,50],index=['H1','H2','H3'])}
df=pd.DataFrame(Data)
print(df)
```

```python
## WAP that selects a particular column from the above data frame, also add new column into it

import numpy as np
import pandas as pd

data={'Kiara':pd.Series([89,90,50],index=['M1','M2','M3']),
    'Meera':pd.Series([80,70,40],index=['M1','M2','M3'])}

df=pd.DataFrame(data)

name=input("Enter the name of the student whose marks u wnt to see:")
print(df[name])

## add a new column
df['shreya']=pd.Series([98,79,80,74],index=['M1','M2','M3','M4'])
print(df)
```

```python
import numpy as np
import pandas as pd

data={'Kiara':pd.Series([89,90,50],index=['M1','M2','M3']),
      'Meera':pd.Series([80,70,40,56],index=['M1','M2','M3','M4']),
      'Saina':pd.Series([10,56,47,90],index=['M1','M2','M3','M4']),
      }

df=pd.DataFrame(data)
print("\nBefore Deletion:")
print(df)
print("\nAfter Deletion:")


## deleting a column
del df['Saina']
print(df)
```
